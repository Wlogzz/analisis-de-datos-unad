```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo
file_path = 'Titanic-Dataset.csv'
data = pd.read_csv(file_path)

# Mostrar las primeras filas del dataset para entender su estructura
data.head(), data.info()

# Configuración para gráficos
sns.set(style="whitegrid")

# Distribución de variables numéricas
numeric_columns = ['Age', 'Fare', 'SibSp', 'Parch']
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    sns.histplot(data[col].dropna(), kde=True, ax=axes[i], color="red")
    axes[i].set_title(f'Distribución de {col}')
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()

# Análisis de correlaciones entre variables numéricas
correlation_matrix = data[['Survived', 'Age', 'Fare', 'SibSp', 'Parch']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Mapa de correlaciones entre variables numéricas")
plt.show()

# Boxplot para detectar valores atípicos en variables clave
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for i, col in enumerate(numeric_columns):
    sns.boxplot(data=data, x=col, ax=axes[i // 2, i % 2], color="lightgreen")
    axes[i // 2, i % 2].set_title(f'Boxplot de {col}')

plt.tight_layout()
plt.show()
```
